find common element between two list using set..

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]


set1 = set(list1)
set2 = set(list2)

common_element = set1.intersection(set2)
print(common_element)
```

remove duplicate words from a sentence using set...

```python
senstence = input("enter the senstence")
words = senstence.split()

unique_words = set(words)
result = " ".join(unique_words)
print(result)
```

```python
list1 = [1,2,3,4,5]
list2 = [4,5,6,7,8]

set1 = set(list1)
set2 = set(list2)

manish = set1 - set2
temp = list(manish)
print(temp)
```

check if two strings are anagrams using set and logic

```python
s1  = input("enter the first string")
s2 = input("enter the second string")
if len(s1) != len(s2):
  print("not anagram")
else:
  is_anagram = True
  for ch in s1:
    if s1.count(ch) != s2.count(ch):
     is_anagram = False
     break
  if is_anagram:
    print("string are anagram")
  else:
    print("string not are anagram")
```
