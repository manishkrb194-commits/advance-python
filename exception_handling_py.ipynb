bank account transaction system

```python
class BankAccount:
    def __init__(self, acc_no, balance):
        self.acc_no = acc_no
        self.balance = balance

    def transfer(self, amount, recipient_acc):
        try:
            if amount > self.balance:
                raise ArithmeticError("Overdraft: Insufficient funds.")

            if not recipient_acc:
                raise NameError("Incorrect account number.")

            if amount > 1000000:
                raise TimeoutError("Transaction timed out due to large volume.")

            self.balance -= amount
            print(f"Successfully transferred ${amount}.")

        except (ArithmeticError, NameError, TimeoutError) as e:
            print(f"Transaction Failed: {e}")

my_acc = BankAccount("123", 500)
my_acc.transfer(1000, "456")
```

student grade system

```python
class StudentGradeSystem:
    def __init__(self):
        self.grades = {}

    def add_grade(self, student_id, grade):
        try:
            if not student_id:
                raise ValueError("Student ID cannot be empty.")

            grade = float(grade)

            if not (0 <= grade <= 100):
                raise ValueError("Grade must be between 0 and 100.")

            self.grades[student_id] = grade
            print(f"Grade for {student_id} updated to {grade}.")

        except ValueError as e:
            print(f"Input Error: {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

sys = StudentGradeSystem()
sys.add_grade("S101", "95")
sys.add_grade("S102", "abc")
```

flightbook system

```python
class FlightBook:
    def __init__(self):
        self.available_seats = 2

    def book_flight(self, name, age):
        try:
            if self.available_seats == 0:
                raise IndexError("No seats available on this flight.")

            if not name or age < 1:
                raise AttributeError("Invalid passenger details provided.")

            if name.lower() == "fail":
                raise RuntimeError("Payment gateway failed.")

            self.available_seats -= 1
            print(f"Booking confirmed for {name}!")

        except (IndexError, AttributeError, RuntimeError) as e:
            print(f"Booking Failed: {e}")

flight = FlightBook()
flight.book_flight("John Doe", 30)
flight.book_flight("Jane Doe", 25)
flight.book_flight("Bob", 40)
```

e commance system

```python
class ECommerce:
    def __init__(self):
        self.inventory = {"Laptop": 5, "Mouse": 0}
        self.coupons = ["SAVE10", "WELCOME"]

    def process_order(self, item, coupon, payment_method):
        try:
            if item not in self.inventory or self.inventory[item] <= 0:
                raise LookupError(f"Error: {item} is out of stock.")

            if coupon not in self.coupons:
                raise PermissionError("Invalid coupon code.")

            if payment_method not in ["Credit", "PayPal"]:
                raise ConnectionRefusedError("Invalid payment method.")

            print(f"Order for {item} processed successfully!")

        except (LookupError, PermissionError, ConnectionRefusedError) as e:
            print(f"Order Processing Error: {e}")

shop = ECommerce()
shop.process_order("Mouse", "SAVE10", "Credit")
```
